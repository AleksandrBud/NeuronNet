```
%tensorflow_version 2.x
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from keras.utils import normalize
```

```
# Загрузим DadaSet
boston = keras.datasets.boston_housing
(train_x, train_y), (test_x, test_y) = boston.load_data()
```

```
df = pd.DataFrame(train_x)
```

```
# Посмотрим на данные
df.describe()
```

```
# Проверим тип данных и есть ли пропуски
df.info()
```

```
# Нормализуем данные
norm_train_x = normalize(train_x)
norm_test_x = normalize(test_x)
```

```
# Создадим слои
model = keras.Sequential()
model.add(keras.layers.Dense(169, activation='relu'))
model.add(keras.layers.Dense(169, activation='relu'))
model.add(keras.layers.Dense(169, activation='relu'))
model.add(keras.layers.Dense(13, activation='relu'))
model.add(keras.layers.Dense(13, activation='relu'))
model.add(keras.layers.Dense(13, activation='relu'))
model.add(keras.layers.Dense(1))
```

```
# Компилируем модель
model.compile(optimizer=tf.keras.optimizers.RMSprop(0.001),
              loss=tf.keras.losses.MeanSquaredError(),
              metrics=['mae', 'mse'])
```

```
# Обучаем
model.fit(norm_train_x, train_y, epochs=10)
```

```
# Смотрим какие показатели выдает нам тестовая выборка
test_acc = model.evaluate(norm_test_x,  test_y, verbose=2)
```

```
# Предсказываем значения тестовой выборки
probability_model = tf.keras.Sequential([model])
predictions = probability_model.predict(norm_test_x)
```

Предсказанные значения более менее похожи на правду. Основное что помогло при обучении модели это использование нескольких метрик.
